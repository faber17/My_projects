Примечание:
<br>
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

Создам сводную таблицу с количеством звонков и их общей продолжительностью. Так как для расчетов важны минуты, округлю значение вверх.

```python
users_calls = pd.pivot_table(calls, index=['user_id', 'month'], values='duration', aggfunc=['count','sum'])
users_calls = users_calls.rename({'count': 'calls_count', 'sum': 'calls_sum'}, axis=1) 
users_calls['calls_sum'] = users_calls['calls_sum'].astype('int') 
display(users_calls)
```

### Помесячная выручка с каждого пользователя

Для начала разделю всех пользователей на два среза по используемому тарифу.

```python
#объединю полученные ранее таблицы
users_all = pd.concat([users_calls, users_messages, users_internet], ignore_index=True, axis=1)
#переименую столбцы
users_all.columns = ['calls_count', 'calls_sum', 'messages_count', 'traffic_sum']
#добавлю для удобства столбец id
users_all['id'] = users_all.index.get_level_values('user_id')
#добавлю для удобства столбец month
users_all['month'] = users_all.index.get_level_values('month')
#добавлю для удобства столбец tariff
users_all = users_all.merge(users[['user_id','tariff']], on='user_id')

tariffs_1 = tariffs.set_index('tariff_name')
display(tariffs_1)

#не все пользователи используют и звонки и смс и интернет. Поэтому NaN заменю на нули, на итоговые значения это не повлияет
users_all['traffic_sum'] = users_all['traffic_sum'].fillna(0)
users_all['messages_count'] = users_all['messages_count'].fillna(0)
users_all['calls_sum'] = users_all['calls_sum'].fillna(0)

def tariff_all(row):
    
    minutes_over = 0
    gb_over = 0
    sms_over = 0

    if row['calls_sum'] > tariffs_1.loc[row['tariff'],'minutes_included']:
        minutes_over = row['calls_sum'] - tariffs_1.loc[row['tariff'],'minutes_included']
    elif row['messages_count'] > tariffs_1.loc[row['tariff'],'messages_included']:
        sms_over = row['calls_sum'] - tariffs_1.loc[row['tariff'],'messages_included']
    elif row['traffic_sum'] > tariffs_1.loc[row['tariff'],'mb_per_month_included']/1024:
        gb_over = row['traffic_sum'] - tariffs_1.loc[row['tariff'],'mb_per_month_included']/1024
        
    return minutes_over * tariffs_1.loc[row['tariff'],'rub_per_minute'] + sms_over * tariffs_1.loc[row['tariff'],'rub_per_message'] + gb_over * tariffs_1.loc[row['tariff'],'rub_per_gb'] + tariffs_1.loc[row['tariff'],'rub_monthly_fee']


#создам столбец с выручкой
users_all['revenue'] = users_all.apply(tariff_all,axis=1)
users_all['revenue'] = pd.to_numeric(users_all['revenue'], downcast='integer')

#проверю результат
display(users_all.head())
display(users_all.tail())
users_all.info()
```

### Гипотеза 1. Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

Для начала проанализирую общую выручку.

```python
all_tariff = users_all['revenue']
all_tariff.hist(bins=100, figsize=(15, 10));
```

Большинство значений (моды) у значений равных абонентской плате.

Посмотрю на среднее, дисперсию и стандартное отклонение выручки по всей выборке и отдельно по тарифам.

```python
users_all_smart = users_all.query('tariff == "smart"')
users_all_ultra = users_all.query('tariff == "ultra"')
```

```python
print("Данные по вcей выборке: среднее", all_tariff.mean().round(2), "дисперсия", all_tariff.var().round(2), "и станд. отклонение", all_tariff.std().round(2))
print("Данные по тарифу smart: среднее", users_all_smart['revenue'].mean().round(2), "дисперсия", users_all_smart['revenue'].var().round(2), "и станд. отклонение", users_all_smart['revenue'].std().round(2))
print("Данные по тарифу ultra: среднее", users_all_ultra['revenue'].mean().round(2), "дисперсия", users_all_ultra['revenue'].var().round(2), "и станд. отклонение", users_all_ultra['revenue'].std().round(2))
```

Нулевая гипотеза, согласно требованиям исследования - "Средняя выручка пользователей тарифов «ultra» и «smart» равна".

Так как я проверяю гипотезу о равенстве средних двух генеральных совокупностей то альтернативаня гипотеза, соответственно - "Средняя выручка пользователей тарифов «ultra» и «smart» различается".

Коэффициент alpha приму равным 0,01, так как это общепринятое значение.

Буду использовать тест Стюдента.

```python
alpha = 0.01

results = st.ttest_ind(users_all_smart['revenue'], users_all_ultra['revenue'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Средняя выручка пользователей тарифов «ultra» и «smart» различается. Это видно даже из простого вычисления среднего по двум выборкам - разница в 1000 рублей.
